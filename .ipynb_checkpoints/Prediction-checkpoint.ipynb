```python
#conversting Date time column from obj to calculable values
matches['date'] = pd.to_datetime(matches['date'])
```

```python
# String to categories then categories to numbers (home and away)
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches['venue'].astype('category').cat.codes
```

```python
matches["hour"] = matches['time'].str.replace(":.+","",regex=True).astype("int")
```

```python
matches['day_code'] = matches['date'].dt.dayofweek
```

```python
matches['target'] = (matches['result'] == 'W').astype('int')
```

```python
matches.head(1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
predictors = ["venue_code","opp_code","hour","day_code"]
```

```python
rf.fit(train[predictors],train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'],preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["target"],prediction=preds))
```

```python
pd.crosstab(index=combined['actual'],columns=combined['prediction'])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"],preds)
```

## Improving precision with rolling averages

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
def rolling_avg(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3,closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ['gf','ga','sh','sot','dist','fk','pk','pkatt']
new_cols = [f"{c}_rolling"for c in cols]
```

```python
new_cols
```

```python
cols
```

```python
rolling_avg(group,cols,new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_avg(x,cols,new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("team")
```
